```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
covid_mobility_us = pd.read_csv('2020_US_Region_Mobility_Report.csv')
```

```python
covid_mobility_us
```

```python
covid_mobility_us = covid_mobility_us.rename(columns={"sub_region_1": "NAME"})
```

```python
covid_mobility_us
```

# CALIFORNIA

```python
california = covid_mobility_us[covid_mobility_us["NAME"] == "California"]

# sns.histplot(data = california, x = 'parks_percent_change_from_baseline')
```

```python
california.drop(columns = ["place_id","iso_3166_2_code"])
```

```python
sns.histplot(california, x = "parks_percent_change_from_baseline")
```

```python
ax_cali = sns.lineplot(x = "date", y = "parks_percent_change_from_baseline", data = california)
# ax.set_xticklabels(ax.get_xticklabels(),rotation = -30)

for ind, label in enumerate(ax_cali.get_xticklabels()):
    if ind % 70 == 0:  # every 14th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
```

# US Dataset

```python
us_grouped_by_date = covid_mobility_us.groupby("date").agg("mean")
# sns.histplot(covid_mobility_us, x = "parks_percent_change_from_baseline")
us_mobility = us_grouped_by_date.reset_index()
us_mobility
```

```python
sns.histplot(us_mobility, x = "parks_percent_change_from_baseline", bins = 50)
plt.title("Parks Percent Change from Baseline in the US")
plt.xlabel("percent change")
```

```python
ax_grouped = sns.lineplot(x = "date", y = "parks_percent_change_from_baseline", data = us_mobility)

for ind, label in enumerate(ax_grouped.get_xticklabels()):
    if ind % 70 == 0:  # every 70th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
        
plt.title("Parks Percent Changes Throughout 2020")
```

```python
# closer look during lockdown


after_03_1 = us_mobility[us_mobility["date"] > "2020-03-01"]
before_june_1 = after_03_1[after_03_1["date"] < "2020-06-02"]
ax_before_june_1 = sns.lineplot(x = "date", y = "parks_percent_change_from_baseline", data = before_june_1)

for ind, label in enumerate(ax_before_june_1.get_xticklabels()):
    if ind % 20 == 0:  # every 20th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
```
